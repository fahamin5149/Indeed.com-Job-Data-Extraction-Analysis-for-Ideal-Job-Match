# Code to extract data about jobs

```python
details2=driver2.find_element(By.ID,'mosaic-jobResults')
details_list=details2.find_elements(By.CLASS_NAME,'css-5lfssm.eu4oa1w0')


new_company=[]
new_title=[]
new_location=[]
new_Gender_Pref=[]
new_education=[]
new_experience=[]
new_joblink=[]
a=1
b=1

for i in details_list:
    try:
        print(a,' ',b)
        boolhelper=False
        jobs_tit=i.find_element(By.CLASS_NAME, 'jobTitle').text
        new_title.append(jobs_tit)
        
        job_loc=i.find_element(By.CLASS_NAME, 'company_location').text
        split_parts = job_loc.split('\n') 
        new_comp = '\n'.join(split_parts[:1])
        new_company.append(new_comp)
        
        new_loc = '\n'.join(split_parts[1:])
        new_location.append(new_loc)
        
        job_link_element = i.find_element(By.CLASS_NAME, 'jcs-JobTitle')
        job_link = job_link_element.get_attribute('href')
        new_joblink.append(job_link)

        new_Gender_Pref.append("No Preference")
        
        job_link_element.click()   #clicking the job link
        time.sleep(10)
        
        job_description_element=driver2.find_element(By.ID,"jobDescriptionText")
        detailsofjob = job_description_element.text


        if "experience" in detailsofjob and "years" in detailsofjob or "year" in detailsofjob or "months" in detailsofjob or "month" in detailsofjob:
            ul_elements = job_description_element.find_elements(By.TAG_NAME, "ul")

            for ul_element in ul_elements:
                li_elements = ul_element.find_elements(By.TAG_NAME, "li")

                for li_element in li_elements:
                    string=li_element.text
                    if("experience" in string and "years" in string or "year" in string or "months" in string or "month" in string):
                        boolhelper=True
                        print(li_element.text)
                        new_experience.append(li_element.text)
                        a=a+1
                        break

        if(boolhelper==False):
            a=a+1
            print("not required")
            notreq="Not required"
            new_experience.append(notreq)
        
        boolhelper=False
        
        if "Graduates" in detailsofjob or "Graduate" in detailsofjob or "bachelor" in detailsofjob or "bachelors" in detailsofjob or "computer science" in detailsofjob or "degree" in detailsofjob or "Bachelor's" in detailsofjob or "bachelor's" in detailsofjob:
            ul_elements = job_description_element.find_elements(By.TAG_NAME, "ul")

            for ul_element in ul_elements:
                li_elements = ul_element.find_elements(By.TAG_NAME, "li")

                for li_element in li_elements:
                    string=li_element.text
                    if "Graduates" in string or "Graduate" in string or "bachelor" in string or "bachelors" in string or "computer science" in string or "degree" in string or "Bachelor's" in string or "bachelor's" in string:
                        boolhelper=True
                        #print(li_element.text)
                        new_education.append(li_element.text)
                        break
        else:
            notreq="Not required"
            new_education.append(notreq)
            boolhelper=True

        if(boolhelper==False):
            notreq="Not required"
            new_education.append(notreq)

        b=b+1
    except NoSuchElementException:
        continue
        
    
    #print(i.text,'\n')
print(new_title)
print(new_company)
print(new_location)
print(new_joblink)
print(new_experience)
```
